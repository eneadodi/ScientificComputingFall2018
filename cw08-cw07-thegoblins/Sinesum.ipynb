````python
import sinesum as s
import numpy as np
import matplotlib.pyplot as plt
````

````python
def allGraphs(x,y,s):
    plt.figure(figsize=(10,6))
    plt.plot(x, y)
    plt.plot(x, y, marker = 'o',color='k')
    plt.title(s)
    plt.xlabel('n values')
    plt.ylabel('Sum values')
    plt.show()
````

````python
n = np.array([1,3,5,10,30,100])
a = np.array([0.01,0.25,0.49])
T = np.multiply(np.pi,2)
y1 = np.array([s.Sn(a[0]*T,i) for i in n])
y2 = np.array([s.Sn(a[1]*T,i) for i in n])
y3 = np.array([s.Sn(a[2]*T,i) for i in n])
allGraphs(n,y1,"Figure with a = 0.01")
y1
allGraphs(n,y2,"Figure with a = 0.25")
y2
allGraphs(n,y3,"Figure with a = 0.49")
y3
````
