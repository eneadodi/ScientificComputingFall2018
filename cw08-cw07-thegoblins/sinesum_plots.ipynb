```python
import numpy as np
import matplotlib.pyplot as plt
import sinesum as ss
```

```python
#N Values
nVal = np.array([1,3,5,10,30,100])
#A values
aVal = np.array([0.01,0.25,0.49])
T = np.multiply(np.pi,2)
j = np.array([1,2,3])
for i in range(3):
    #Creates all the necessary plots
    f = plt.figure(figsize =(8,6))
    a = plt.axes()
    yVals = np.array([ss.Sn(aVal[i]*T,a) for a in nVal])
    a.plot(nVal,yVals,marker = "o", color = "blue")
    a.set(xlabel= "iterations/n values: ", ylabel= ("a ="+str(aVal[i])), title="Sine sum "+str(j[i]))
    plt.show()
    print(yVals)
```

```python
#N Values
nValue = 100
T = 2*np.pi
for i in range(3):
    #Creates all the necessary plots
    f = plt.figure(figsize=(6,4))
    a = plt.axes()
    yVals = np.array([ss.Sn(aVal[i]*T,a+1) for a in range(nValue)])
    a.plot(np.arange(0,nValue,1), yVals, color = "blue")
    a.set(xlabel= "iterations (n): "+str(nValue), ylabel= ("a ="+str(aVal[i])), title="Sine sum")
    plt.show()
```

```python
extraA = np.array([0.23,0.27])
for i in range(2):
    #Creates all the necessary plots
    f = plt.figure(figsize =(8,6))
    a = plt.axes()
    yVals = np.array([ss.Sn(extraA[i]*T,a) for a in nVal])
    a.plot(nVal,yVals,marker = "o", color = "blue")
    a.set(xlabel= "iterations/n values: ", ylabel= ("a ="+str(extraA[i])), title="Sine sum "+str(j[i]))
    plt.show()
    print(yVals)
```

**On top of the required graphs, There is a section in this notebook which contains 3 graphs and they show how sinesum approximates the values. Specficially, it shows the amount of oscillations involved in the approximation of the final y-value in regards to $\alpha$. Once again, we see when $\alpha$ is 0.01 or 0.49, we get the same graph, but when $\alpha$ is 0.25, we get a tremendous amount of oscillations (damping) to the value of 1. This seems like it would be the most accurate representation of the piecewise function discussed:**
  $$f(t) = \begin{cases}
             1, & 0 < t < T/2 \\
             0, & t = 0 \\
             -1, & -T/2 < t < 0
           \end{cases}$$
