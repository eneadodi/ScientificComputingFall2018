# Taylor Series
##### The Taylor series is one of math's most powerful tools for appoximating functions!
Lets say there is a function $f(x)$ where we do not know the value of $x$, but rather a value $a$. We can say that we can approximate the value of f(x) with the value of f(a). This however is rarely the case because the approximation is constant. f(a) is known as the 0th order approximation. The next step in finding the value of f(x) would be to think of the f(x) equation as a line rather than a constant number. We know the slope of a line is:
$$f'(x) = \frac{f(x)-f(a)}{x-a}$$
and we can re-arrange that function to get the value of f(x)
$$f(x) = f(a) + f'(a)(x-a)$$
At this point we have a linear approximation of f(x) also known  as a 1st order approximation. We can continue approximating the value at f(x) better and better by increasing the order of approximation. The 2nd order appoximation would then be based on a parabola and can be gotten by getting the integral of f'(x)
$$f(x) = \int_{}{}f'(x)d(x-a) = f'(a)(x-a)+\frac{f''(a)(x-a)^2}{2}$$
thus now we have a 2nd order approximation:
$$f(a) + f'(a)(x-a)+\frac{f''(a)(x-a)^2}{2}$$
We can repeat this process as many times as necessary to get better and better approximations!

```python
import taylor_approx as ta
import sympy as sp
import numpy as np
```

```python
sp.init_printing()  # Tell sympy to use the prettiest printing available (in Jupyter, this means $\LaTeX$)
x,y,z,t,w,k = sp.symbols('x y z t w k')         # Define common variable names to be mathematical variable symbols
f,g,e = sp.symbols('f g e', cls=sp.Function)  # define variables f and g to be mathematical function symbols
f = sp.sin(x)
g = sp.tanh(x)
e = (x**2)/10 + sp.sin(2*x)/2
xVals = np.linspace(-5,5,1000)
n = np.array([0,1,2,3,4,5])
a = np.array([-3,0,2])
```

### f = sin(x)

```python
name = "sine taylor"
for i in a:
    for j in n:
        d = ta.derivative(f,x,i,j)
        taylorExpansion = ta.taylor(xVals,i,d)
        ta.plot(taylorExpansion,xVals,i,j,name)
```

### f = tanh(x)

```python
name = "tanh(x)"
for i in a:
    for j in n:
        d = ta.derivative(g,x,i,j)
        taylorExpansion = ta.taylor(xVals,i,d)
        ta.plot(taylorExpansion,xVals,i,j,name)
```

$$f = (x^2)/10 + sin(2x)/2$$

```python
name = "(x^2)/10 + sin(2x)/2"
for i in a:
    for j in n:
        d = ta.derivative(e,x,i,j)
        taylorExpansion = ta.taylor(xVals,i,d)
        ta.plot(taylorExpansion,xVals,i,j,name)
```
