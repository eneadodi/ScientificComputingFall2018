$$f'(x) = \frac{df}{dx}(x) = Df(x) \equiv \lim_{h \to 0} \frac{f(x+h) - f(x)}{(x + h) - x} = \lim_{h\to 0} \frac{f(x) - f(x - h)}{x - (x - h)} = \lim_{h \to 0} \frac{f(x+h) - f(x-h)}{(x+h)-(x-h)}$$

As you can see, the derivative of a function can be calculated in multiple ways. One of the ways is by using a differential operator known as the gradient. The gradient is simply a n*n matrix with two diagonals: An example of this gradient with n = 6 is:
|      |      |      |      |     |     |
|------|------|------|------|-----|-----|
| -1   | 1  | 0    | 0    | 0   | 0   |
| -0.5 | 0    | 0.5  | 0    | 0   | 0   |
| 0    | -0.5 | 0    | 0.5  | 0   | 0   |
| 0    | 0    | -0.5 | 0    | 0.5 | 0   |
| 0    | 0    | 0    | -0.5 | 0   | 0.5 |
| 0    | 0    | 0    | 0    | -1  | 1   |

Thus, the gradient function creates a matrix such as the one above for an array of x values n.
The gradient is important for the calculation of forward, backward, and central differences. Again, these equations are as follow:
$$\lim_{h \to 0} \frac{f(x+h) - f(x)}{(x + h) - x} = \lim_{h\to 0} \frac{f(x) - f(x - h)}{x - (x - h)} = \lim_{h \to 0} \frac{f(x+h) - f(x-h)}{(x+h)-(x-h)}$$
Thus, we divide the values in each of the cells of the matrix by dx to get the difference formulas. The first two cells and the last two cells are -1,1 because central differencet cannot be calculated for them, thus we have to use forwards and backwards differences.
Finally, we do matrix multipliation using the y values calcuated by f(x) and the gradient to get the y values of the derivative!
