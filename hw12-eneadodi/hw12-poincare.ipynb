```python
@nb.jit
def scatterhw12(F = 0.4,dt = 0.001,o = 0):
    '''scatter method
    '''
    oTitle = str(o*100)
    odes = s.solve_odes(0,0,F,T = 1000*np.pi)
    for i in range(1000):
        plt.scatter(odes["x"][(int)(o+(i/dt)*2*np.pi)],odes["y"][(int)(o+(i/dt)*2*np.pi)], marker='.', color = 'k')
    plt.set_xlim(-1.5,1.5)
    plt.set_ylim(-1.1,1.1)
    plt.set_xlabel("x")
    plt.set_ylabel("x'")
    plt.savefig('frame'+oTitle+'.png')

@nb.jit
def allScatters(F = 0.4,dt = 0.001):
    os = np.arange(0,1,0.01)
    for i in os:
        scatterhw12(F,dt,i)
```
