```python
import sombrero as s
import numpy as np
```

```python
xVal = -0.9
yVal = 0
F = 0.18
odePoints = s.solve_odes(xVal,yVal,F)
s.plotxy(odePoints,("Parametric Curve, F =" + str(F)+ ", Inital x =" + str(xVal) + ", Inital y =" + str(yVal)))
s.plotx(odePoints, "Position vs Time Graph")
s.scatter(odePoints,F,"Poincare section")
```

**interpretation** The parametric graph shows a circular pattern. If i was to guess what this means in regards to the system, it might represent the ball oscillating left and right over a hump on the sombrero. The x vs t graph shows exactly what i expected, the ball's x position change in time and its oscilations. The peaks where it is more curved (where the tangent = 0) probably represents the slight moment when the x values are not changing. The Poincare section dosen't give me any useful information.

```python
F = 0.4
xVal = 0
yVal = 0
odePoints = s.solve_odes(xVal,yVal,F, T = 1000*np.pi)
s.plotxy(odePoints,("Parametric Curve, F =" + str(F)+ ", Inital x =" + str(xVal) + ", Inital y =" + str(yVal)))
s.plotx(odePoints,"Position vs Time Graph")
s.scatter(odePoints,F,"Poincare section",N=500)
```

**interpretation** The structure of the Poincare section now is a curve (I changed N to 10* the normal value) where we can actually see the value not only increasing, but decreasing after a certain point (x=0.8). The parametric curve now is extremeley interesting because the "two entities" and their connections to eachother have become more detailed with the increase of range in t. Also, if you compare the centers of both sides, It seems like where there is lines on the left, there aren't on the right, and vice versa. Overall, I believe the parametric curve simply demostrates the oscillating pattern of the "ball" on the sombrero
